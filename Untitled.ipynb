```python
import os, sys
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing import image


def load_batch_images(batch_size, dirName):
    image_names = np.array(os.listdir(os.path.join(dirName)))
    index = np.random.randint(0, image_names.shape[0], batch_size)
    image_names = image_names[index]
    imgs = []
        
    for i in range(len(image_names)):
        img = image.load_img(os.path.join(dirName, image_names[i]), target_size=(96, 96))
        imgs.append(img)
    return imgs
            
imgs = load_batch_images(9, 'image_raw/')
row = 3
col = 3

fig, axs = plt.subplots(row, col, figsize = (6, 6))
count = 0
for i in range(row):
    for j in range(col):
        axs[i, j].imshow(imgs[count])
        axs[i, j].axis('off')
        count += 1
```
